```python
# 불 자료형
참 = True
거짓 = False
print(참, type(참))
print(거짓, type(거짓))
print(100 == 20)
print(10 > 5)
```

```python
#문자열 자료형

a = '100'
b = 'Python'

#문자열 연산

print(a+b) #문자열과 문자열을 연결
print(a*10) #특정 횟수만큼 반복
print(a < b) #문자의 길이 비교
print('1' in a) #특정 문자열 포함하는지
print('1' not in a)#특정 문자열 포함하지 않는지
```
